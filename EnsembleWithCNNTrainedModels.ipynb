```python
import pandas as pd
```

```python
cnn1 = pd.read_csv('./test_resultCNN.csv')
```

```python
cnn2 = pd.read_csv('./test_resultCNN2.csv')
```

```python
imgAugCnn1 = pd.read_csv('./test_resultCNNImageAugmentation.csv')
```

```python
imgAugCnn2 = pd.read_csv('./test_resultCNNImageAugmentation2.csv')
```

```python
nn = pd.read_csv('./test_resultNN.csv')
```

```python
df = cnn1.merge(cnn2, how='inner', on='ImageId', suffixes=('1','2'))
```

```python
df = df.merge(cnn2, how='inner', on='ImageId')
```

```python
df = df.merge(imgAugCnn1, how='inner', on='ImageId', suffixes=('3','4'))
```

```python
df = df.merge(imgAugCnn2, how='inner', on='ImageId')
```

```python
df = df.merge(nn, how='inner', on='ImageId', suffixes=('5','6'))
```

```python
df.head()
```

```python
def foo(count):
    maxim = 0
    val = -1
    equalFound = False
    for k in count:
        if maxim == count[k]:
            equalFound = True
        if count[k] > maxim:
            maxim = count[k]
            val = k 
            equalFound = False
    return val, equalFound
```

```python
pred = [] 
idx = []
from collections import Counter
for i in range(df.shape[0]):
    c = Counter(df.iloc[i][['Label1', 'Label2', 'Label3', 'Label4', 'Label5', 'Label6' ]])
    v, eq = foo(c)
    if eq:
        pred.append(df.iloc[i]['Label5'])  #giving weightage to fifth model
    else:
        pred.append(v)
    idx.append(df.iloc[i]['ImageId'])
```

```python
res = pd.DataFrame()
res['ImageId'] = idx
res['Label'] = pred
res[['ImageId', 'Label']].to_csv('ensemble.csv', index=False)
```
